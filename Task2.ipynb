```python
# Function to return the dataframe having same city as provied by user
def find_all_res_of_city(city_name,df,vectors_city):
    city_name = city_name.replace(' ','')  # removing space
    num_of_res = df[df['City']==city_name].shape[0]  # finding number of rows/data having same city as given user
    if num_of_res == 0:   # if city not match by our dataset then return empty dataframe
        return df.drop(columns = df.columns)
    user_vector_city = cv.transform([city_name])  # converting user city into vector form
    user_vector_city = user_vector_city.toarray()
    similar = cosine_similarity(user_vector_city,vectors_city)  # finding the most similar city with respect to the given city
    most_similar_index = similar[0].argsort()[-num_of_res:][::-1]  # getting the index of similar city and return all of them
    df_city = df.iloc[most_similar_index]
    return df_city
```

```python
def recommend_res(city_name,cuisine_preference1,cuisine_preference2,avg_cst_fr_2,agg_rating,df,vectors_city):
    df_city = find_all_res_of_city(city_name,df,vectors_city)  # get dataframe of all city similar to given city by user
    if df_city.shape[1] == 0:  # if dataframe is empty then return "City not found"
        return "City Not Found"
    # extracting our features which is cuisines,avg cost for two and aggregate rating
    df_city['features'] = df_city['Cuisines'] + " " + df_city['Average Cost for two'].astype(str) + " " + df_city['Aggregate rating'].astype(str);
    df_city['features'] = df_city['features'].apply(lambda x : x.replace(', ',' '))
    cv2 = CountVectorizer(max_features = 10000,stop_words = 'english')
    vectors = cv2.fit_transform(df_city['features']).toarray()  # converting into vector form
    similarity2 = cosine_similarity(vectors)
    user_input_cus = cuisine_preference1 + " " + cuisine_preference2 + " " + str(avg_cst_fr_2) + " " + str(agg_rating)
    user_input_cus_vec = cv2.transform([user_input_cus])  # converting user input to vector form
    user_input_cus_vec = user_input_cus_vec.toarray()    
    similar2 = cosine_similarity(user_input_cus_vec,vectors)  # finding similarity b/w dataset and given input
    most_similar_index2 = similar2[0].argsort()[-20:][::-1]   # returning top 20 similar index
    df_city = df_city.iloc[most_similar_index2]
    return df_city.drop(columns = ['feature_city','features'])[0:10]  # we return only top 10 restaurants detail similar to given user input
```

```python
# taking user input
city_name = input("Enter City Name: ").lower()
cuisine_preference1 = input("Enter Cuisines Preference 1: ").lower()
cuisine_preference2 = input("Enter Cuisines Preference 2: ").lower()
avg_cst_fr_2 = int(input("Enter Avg Cost for two: "))
agg_rating = float(input("Aggregate Rating: "))
```

```python
# Recommend the top 10 restaurants

recommend_res(city_name,cuisine_preference1,cuisine_preference2,avg_cst_fr_2,agg_rating,df,vectors_city)
```
